## Project_2: Random Steps

### Importing Libraries

```python
import matplotlib.pyplot as plt
from random import choice
%matplotlib inline  
# %matplotlib inline --> for Visualization to pop-up in Jupyter Notebook
```

### -------------------------------------------------------------------------------------------------------------------

```python
plt.rc('figure',figsize=(8,5))     # change default settings for plots
```

```python
step_option=[1,-1]
step_choice=choice(step_option)       # getting random numbers from -1 or 1 randomly
step_choice
```

```python
# Function_1
def random_walk(step_num):          # a function to generate steps 1 or -1 randomly based on given number
    walk=[]
    step_choice=choice([1,-1])
    walk.append(step_choice)
    for i in range(1,step_num):
        step_choice_2=choice([1,-1])
        next_step=walk[i-1]+ step_choice_2                  # if 100, so 1-1  then either +1  or + -1
        walk.append(next_step)
    return walk  
```

```python
rand_w=random_walk(1000)            # calling out function with number : 10000
print(rand_w)
print(plt.plot(rand_w))
```

```python
# Function_2
def plot_walk(walk):     # a function that will plot given number with labels for Y-axis,X-axis, and Title 
    plt.plot(walk)
    plt.xlabel('Number of Steps')
    plt.ylabel('Distance from Origin')
    plt.title('Our Random Walk')
    plt.show()
```

```python
plot_walk(random_walk(1000))     # Function_1 inside Function_2 with given number: 1000
```

### -----------------------------------------------------------------------------------------------------------

```python
def NewPlot_walks(num,walk_steps):
    lab_list=list(range(1,num+1))
    for i in range(0,num):
        x=list(range(1,walk_steps+1))
# using Function_1 
        plt.plot(x,random_walk(walk_steps),label='Plot Number'+' '+str(lab_list[i]))
        plt.xlabel('Number of Steps')
        plt.ylabel('Distance from Origin')
        plt.title('Random Walk')
        plt.legend(loc='lower left')
    plt.show() 
```

```python
NewPlot_walks(2,1000)
```
