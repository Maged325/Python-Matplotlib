### Loading Libraries

```python
from PIL import Image,ImageFilter,ImageFont, ImageDraw
from matplotlib.pyplot import imshow
# PIL : Python Image Library
```

### Loading Image

```python
filename=(r'D:\My Work\Data Analyst\Python\Python_Notes\Matplotlib_Files\Datasets\forest.jpg')
im=Image.open(filename)
imshow(im)
```

#### Getting Information about picture

```python
print(im.size)
print(im.format_description)
print(im.mode)
```

#### Applying some Filters

```python
crop=(200,200,600,600)
cropped_image=im.crop(crop)
imshow(cropped_image)
```

```python
resized_im=im.resize((300,200))
imshow(resized_im)
```

```python
gray_scale=im.convert('LA')     # luminosity (brightness) and alpha
imshow(gray_scale)
```

```python
effect=im.filter(ImageFilter.GaussianBlur(radius=20))
imshow(effect)
```

```python
draw=ImageDraw.Draw(im)
font=ImageFont.truetype("arial.ttf",100)
draw.text((70,50),"Nature",font=font)
imshow(im)
```
