## **Dual-Line Plot with Markers and Labels:** Plot the revenue and profit for a company over six months on the same graph. Use different markers for each line and ensure the plot has labeled axes, a title, and a legend. Ensure the markers are distinguishable and the lines have different styles.

```python
# Importing libraries
import matplotlib.pyplot as plt

# Sample data for six months, along with their revenue and profit
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
revenue = [10000, 15000, 18000, 16000, 12000, 20000]
profit = [3000, 4500, 3500, 3200, 3800, 4200]

# Create a figure and axis, to plot multiple plots on a single figure
fig, ax1 = plt.subplots()

# Plot first line plot as 'revenue' with different marker and linestyle
ax1.plot(months, revenue, color='tab:blue', marker='o', linestyle='-', label='Revenue')
ax1.set_xlabel('Month')
ax1.set_ylabel('Revenue', color='tab:blue',fontdict={'fontsize':16})
ax1.tick_params(axis='y', labelcolor='tab:blue')

# Create a second y-axis to plot profit
ax2 = ax1.twinx()     # create invisible X- axis and independent Y-axis
ax2.plot(months, profit, color='tab:green', marker='s', linestyle='--', label='Profit')
ax2.set_ylabel('Profit', color='tab:green',fontdict={'fontsize':16})
ax2.tick_params(axis='y', labelcolor='tab:green')

# Adding title and show grid
plt.title('Company Revenue and Profit Over Six Months')
ax1.grid(True)

# Adding legends
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

# Show plot and automatically adjusts subplot parameters so that the subplot(s) fits in to the figure area.
plt.tight_layout()
plt.show()
```

## **Vertical Bar Chart with Annotations:** Create a vertical bar chart showing the number of books read by students in five different grades. Annotate the top of each bar with the exact number of books read. Ensure the bars are colored differently and labeled.

```python
import matplotlib.pyplot as plt

# Data for the vertical bar chart, with different colours
grades = ['Grade 1', 'Grade 2', 'Grade 3', 'Grade 4', 'Grade 5']
books_read = [30, 45, 60, 40, 50]
colors = ['blue', 'orange', 'green', 'red', 'purple']

# Creating the bar chart
fig, ax = plt.subplots()
bars = ax.bar(grades, books_read, color=colors)

# Annotate each bar with the number of books read
for bar in bars:
    height = bar.get_height()  # Get the height of the bar
    # Label the bar with the height value
    ax.text(
        bar.get_x() + bar.get_width() / 2,  # X position (center of the bar)
        height,                             # Y position (top of the bar)
        str(height),                        # Text (number of books read)
        ha='center',                        # Center-align horizontally
        va='bottom'                         # Place text just above the bar
    )

# Labeling the Y-axis, X-axis, and title
ax.set_xlabel('Grades',fontdict={'fontsize':14})
ax.set_ylabel('Number of Books Read',fontdict={'fontsize':14})
ax.set_title('Number of Books Read by Students in Different Grades',fontdict={'fontsize':16})

# Show the plot and automatically adjusts subplot parameters so that the subplot(s) fits in to the figure area.
plt.tight_layout()
plt.show()
```

## **Customized Pie Chart:** Create a pie chart showing the market share of five smartphone brands. Ensure the slices have different colors, labels with percentages, and explode the slice with the highest market share. Also, rotate the chart to start from a different angle.

```python
import matplotlib.pyplot as plt

# Data for different smartphone brands for the pie chart, with different colours for each slice.
brands = ['Samsung', 'Huawei', 'Xiaomi', 'Apple', 'Lenovo']
market_share = [25, 15, 30, 10, 20]
colors = ['red','orange','green','yellow','blue']

# Explode the slice with the highest market share
explode = [0.1 if share == max(market_share) else 0 for share in market_share]
# alternative way to use 'explode' to slice with the highest market share
# explode=(0, 0, 0.4, 0, 0) 

# Create the pie chart
fig, ax = plt.subplots()
wedges, texts, autotexts = ax.pie(market_share, explode=explode , labels=brands, colors=colors,
    autopct='%1.1f%%',  # Format for the percentage labels
    startangle=140,     # Rotate the pie chart
    shadow=True )       # Add shadow for better visual effect

# Customize the font size of the labels and percentages
for text in texts:
    text.set_fontsize(10)
for autotext in autotexts:
    autotext.set_fontsize(10)

# Set the title
ax.set_title('Market Share of Smartphone Brands')

# Show the plot
plt.show()
```

## **Stacked Bar Chart with Labels:** Create a stacked bar chart showing the performance of three teams in four different tasks. Each team’s performance in each task should be stacked on top of the others. Add labels inside the bars to indicate the performance numbers, and customize the colors.

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the stacked bar chart
tasks = ['Task 1', 'Task 2', 'Task 3', 'Task 4']
teams = ['Team A', 'Team B', 'Team C']
performance = {'Team A': [10, 15, 20, 25], 'Team B': [12, 18, 25, 30],'Team C': [8, 12, 18, 22]}

# Convert the performance data into arrays
performance_data = np.array([performance[team] for team in teams])

# Colors for each team's performance
colors = ['#ff9999', '#66b3ff', '#99ff99']

# Create the figure and axis
fig, ax = plt.subplots()

# Initialize bottom for stacking bars
bottoms = np.zeros(len(tasks))

# Plot each team's performance
for i, team in enumerate(teams):
    ax.bar(tasks, performance_data[i], bottom=bottoms, color=colors[i], label=team)
    # Update bottoms to the current stack
    bottoms += performance_data[i]

    # Adding labels inside bars
    for j, value in enumerate(performance_data[i]):
        # Calculate the position for the label (middle of the bar)
        ax.text(
            j, 
            bottoms[j] - value / 2, 
            str(value), 
            ha='center', 
            va='center', 
            color='black' )

# Labeling the axes and the plot
ax.set_xlabel('Tasks')
ax.set_ylabel('Performance')
ax.set_title('Team Performance in Various Tasks',fontdict={'fontsize':14})
ax.legend()

# Show the plot
plt.tight_layout()
plt.show()
```

## Overlapping Line Plots with Annotations: Plot three trigonometric functions (sine, cosine, and tangent) on the same graph over the range of 0 to 2π. Ensure each line has a different style and color. Add annotations for key points where the functions intersect or reach significant values (e.g., sine and cosine crossing at π/4). Include a legend and grid.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range
x = np.linspace(0, 2 * np.pi, 1000)

# Compute function values
sin_x = np.sin(x)
cos_x = np.cos(x)
tan_x = np.tan(x)

# Plot the functions
plt.figure(figsize=(10, 6))

# Plot sine function
plt.plot(x, sin_x, label='sin(x)', color='blue', linestyle='-', linewidth=2)

# Plot cosine function
plt.plot(x, cos_x, label='cos(x)', color='red', linestyle='--', linewidth=2)

# Plot tangent function
plt.plot(x, tan_x, label='tan(x)', color='green', linestyle=':', linewidth=2)

# Add grid
plt.grid(True)

# Annotate intersection of sin(x) and cos(x) at x = π/4
x_intersection = np.pi / 4
y_intersection = np.sin(x_intersection)  # sin(x) = cos(x) at this point
plt.annotate(
    'sin(x) = cos(x)\n(π/4, √2/2)', 
    xy=(x_intersection, y_intersection), 
    xytext=(x_intersection + 0.5, y_intersection + 0.5),
    arrowprops=dict(facecolor='black', shrink=0.05),
    fontsize=12
)

# Annotate where tan(x) crosses y-axis
plt.annotate(
    'tan(x) at x = π/2 (Asymptote)', 
    xy=(np.pi / 2, np.tan(np.pi / 2)), 
    xytext=(np.pi / 2 + 0.5, 10), 
    arrowprops=dict(facecolor='black', shrink=0.05),
    fontsize=12
)

# Add legend
plt.legend()

# Set labels and title
plt.xlabel('x')
plt.ylabel('Function value')
plt.title('Trigonometric Functions: sin(x), cos(x), and tan(x)')

# Set limits for y-axis to avoid extreme values of tan(x)
plt.ylim(-2, 2)

# Show plot
plt.show()
```

## **Advanced Subplots with Different Scales:** Create a figure with three subplots: the first subplot shows a linear function, the second shows an exponential function, and the third shows a logarithmic function. Use different y-axis scales (linear, log, and symlog) for each subplot. Add a grid to all subplots and customize the lines and markers.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define x range
x = np.linspace(0, 10, 500)

# Linear function
linear_y = 2 * x + 1

# Exponential function
exp_y = np.exp(x)

# Logarithmic function
log_y = np.log(x + 1)  # Adding 1 to avoid log(0)

# Create a figure and axes
fig, axs = plt.subplots(3, 1, figsize=(10, 15))

# Plot linear function on the first subplot
axs[0].plot(x, linear_y, color='blue', linestyle='-', marker='o', markersize=4)
axs[0].set_title('Linear Function')
axs[0].set_xlabel('x')
axs[0].set_ylabel('y')
axs[0].grid(True)

# Plot exponential function on the second subplot with log scale
axs[1].plot(x, exp_y, color='red', linestyle='--', marker='x', markersize=5)
axs[1].set_title('Exponential Function')
axs[1].set_xlabel('x')
axs[1].set_ylabel('y')
axs[1].set_yscale('log')
axs[1].grid(True, which='both', linestyle='--', linewidth=0.7)

# Plot logarithmic function on the third subplot with symlog scale
axs[2].plot(x, log_y, color='green', linestyle='-.', marker='^', markersize=6)
axs[2].set_title('Logarithmic Function')
axs[2].set_xlabel('x')
axs[2].set_ylabel('y')
axs[2].set_yscale('symlog', linthresh=0.1)
axs[2].grid(True)

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

## **Complex Bar Chart with Custom Legends and Colors:** Create a horizontal grouped bar chart showing the test scores of four different students in five subjects. Each group of bars should represent one subject with bars colored differently for each student. Add a custom legend, and ensure the bars are labeled with their respective scores and ordered by the highest to lowest scores for each subject.

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the bar chart
subjects = ['Math', 'Science', 'English', 'History', 'Art']
students = ['Alice', 'Bob', 'Charlie', 'David']

# Test scores for each student in each subject
scores = {
    'Alice': [85, 90, 78, 88, 95],
    'Bob': [80, 85, 88, 85, 92],
    'Charlie': [70, 75, 80, 78, 85],
    'David': [90, 92, 85, 90, 88]
}

# Colors for each student
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728']

# Number of subjects and students
num_subjects = len(subjects)
num_students = len(students)

# Calculate bar width and positions
bar_width = 0.15
index = np.arange(num_subjects)

# Create the figure and axes
fig, ax = plt.subplots(figsize=(12, 8))

# Create bars for each student
for i, student in enumerate(students):
    # Calculate the position for each student's bars
    bar_positions = index + i * bar_width
    ax.barh(bar_positions, scores[student], bar_width, color=colors[i], label=student)

    # Label the bars with scores
    for pos, score in zip(bar_positions, scores[student]):
        ax.text(score + 1, pos, f'{score}', va='center')

# Set the labels and title
ax.set_yticks(index + bar_width * (num_students - 1) / 2)
ax.set_yticklabels(subjects)
ax.set_xlabel('Score')
ax.set_title('Test Scores of Students by Subject')
ax.legend(title='Students')
ax.grid(True, linestyle='--', alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```
