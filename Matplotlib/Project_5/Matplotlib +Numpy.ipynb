# using NumPy array to plot graphs 

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
xpoints=np.array([1,2,3,4,8])
ypoints=np.array([3,5,6,7,10])

plt.plot(xpoints,ypoints,'o',ms='10')    # '*', '.', 'x'
plt.show()
```

```python
xpoints=np.array([1,2,3,4,8])
ypoints=np.array([3,5,6,7,10])

plt.plot(xpoints,ypoints,'o:r',mec='r',mfc='b',linewidth='4',markersize='10')    # 'marker-line-colour   
# mfc: to set colour inside marker
# mec: to set colour of the edge marker
# linstyle= 'dashed'
plt.show()
# line style: '-', ':', '-.', '--'
```

```python
xpoints=np.array([1,2,3,4,8])
ypoints=np.array([3,5,6,7,10])

font1= {'family':'serif','color':'blue','size':20}
font2={'family':'serif','color':'darkred','size':15}

plt.plot(xpoints,ypoints,'o:r',mec='r',mfc='b',linewidth='4',markersize='10')   
# plt.subplot(1,2,2)

plt.title('New Title Here',fontdict=font1,loc='left')
plt.suptitle('new subtitle')
plt.ylabel('Y-axis',fontdict=font2)
plt.xlabel('X-axis',fontdict=font2)

plt.grid(axis='x',color='green',linestyle='--',linewidth=1)
# plt.grid()
# plt.grid(axis='y')

plt.show()
```

```python
x= np.array(['A','B','C','D'])
y= np.array([3,8,1,10])
plt.barh(x,y)
plt.show()
```

```python
x= np.array(['A','B','C','D'])
y= np.array([3,8,1,10])
plt.bar(x,y,width=0.5)
plt.show()
```
